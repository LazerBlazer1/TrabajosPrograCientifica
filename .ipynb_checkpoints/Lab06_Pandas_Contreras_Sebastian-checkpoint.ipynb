```python
#P2.1: Faltan 20 datos NaN.
#P2.2: Para demostrar si temperatura maxima y mínima con un 19% superior, la importancia de esto básicamente rádica en el tipo de limpieza que deba hacerse al 
#dataset, siendo más efectivo el rellenar los NaN o directamente eliminar las filas con datos corruptos
#P2.3: Porque creo que es un resultado más directo, dado que si entregan 100 datos nulos como tal en total, no me daría una respuesta significativa a si en realidad que método usar, en cambio si 
#me entregaran un "10%" me entregaría una cifra significativa para poder utilizar una limpieza efectiva sobre el dataset.
```

```python
#Ejercicio 3
df_limpio = df.copy()
df_limpio
```

```python
df_limpio['temp_max'] = df_limpio['temp_max'].fillna(df_limpio['temp_max'].mean())
df_limpio['temp_min'] = df_limpio['temp_min'].fillna(df_limpio['temp_min'].mean())

df_limpio['humedad_pct'] = df_limpio['humedad_pct'].fillna(df_limpio['humedad_pct'].median())

df_limpio['precipitacion_mm'] = df_limpio['precipitacion_mm'].fillna(0.0)

filas_antes = len(df_limpio)
df_limpio.dropna(subset=['condicion'], inplace=True)
filas_despues = len(df_limpio)

print("Resumen de nulos después de la limpieza:")
print(df_limpio.isnull().sum())
```

```python
limpieza = df_limpio.isnull().sum()
print(limpieza)
df_limpio
```

```python
#Preguntas de reflexion
#P3.1: mean se utiliza para datos simetricos sin valores extraños, median para datos muy sesgados y valores atipicos como tal
#P3.2: Dos filas de NaN fueron dropeadas, es aceptable debido a que es menor la significancia de perdida de filas para
#Un mejor analisis
#P3.3: Se pierde gran parte de los datos dados, quedando demasiado pocos para un analisis util de algun segmento de investigacion.
#No es factible sin un tratamiento de datos significativo.
a = df.dropna()
print(a)
```
