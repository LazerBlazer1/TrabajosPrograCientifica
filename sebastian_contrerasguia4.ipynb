```python
#Ejercicio 1
class Empleado:
    """Clase base que representa un empleado genérico."""
    def __init__(self, nombre, empresa, salario_base):
        self.nombre = nombre
        self.empresa = empresa
        self.salario_base = salario_base

    def calcular_salario(self):
        """Calcula el salario del empleado."""
        return self.salario_base

    def descripcion(self):
        """Retorna una descripción general del empleado."""
        # Se formatea el número para que tenga espacios como en la salida esperada
        salario_formateado = f"{self.calcular_salario():,}".replace(',', ' ')
        return f"Empleado: {self.nombre} - Salario: ${salario_formateado}"

class GerenteVentas(Empleado):
    """Subclase para un gerente de ventas."""
    def __init__(self, nombre, empresa, salario_base, bono_ventas):
        super().__init__(nombre, empresa, salario_base)
        self.bono_ventas = bono_ventas

    def calcular_salario(self):
        """Calcula el salario sumando el bono al salario base."""
        return self.salario_base + self.bono_ventas

    def descripcion(self):
        """Retorna la descripción específica del gerente."""
        salario_formateado = f"{self.calcular_salario():,}".replace(',', ' ')
        return f"Gerente Ventas: {self.nombre} - Salario: ${salario_formateado}"

class ObreroPorHora(Empleado):
    """Subclase para un obrero pagado por hora."""
    def __init__(self, nombre, empresa, horas_trabajadas, valor_hora):
        # El obrero no tiene un salario base fijo, le pasamos 0 a la clase padre
        super().__init__(nombre, empresa, 0)
        self.horas_trabajadas = horas_trabajadas
        self.valor_hora = valor_hora

    def calcular_salario(self):
        """Calcula el salario multiplicando horas por valor hora."""
        return self.horas_trabajadas * self.valor_hora

    def descripcion(self):
        """Retorna la descripción específica del obrero."""
        salario_formateado = f"{self.calcular_salario():,}".replace(',', ' ')
        return f"Obrero PorHora: {self.nombre} - Salario: ${salario_formateado}"

# Pruebas para validar la salida esperada
gerente = GerenteVentas("Ana Pérez", "Empresa X", 2000000, 300000)
obrero = ObreroPorHora("Pedro Soto", "Empresa Y", 160, 4000)

print(gerente.descripcion())
print(obrero.descripcion())
```

```python
#ejercicio 2
class Electrico:
    """Clase mixin para vehículos eléctricos."""
    def __init__(self, bateria_kwh, **kwargs):
        super().__init__(**kwargs)
        self.bateria_kwh = bateria_kwh

class Combustion:
    """Clase mixin para vehículos de combustión."""
    def __init__(self, cilindrada_cc, **kwargs):
        super().__init__(**kwargs)
        self.cilindrada_cc = cilindrada_cc

class Vehiculo:
    """Clase base para vehículos."""
    def __init__(self, marca, modelo, año, **kwargs):
        super().__init__(**kwargs)
        self.marca = marca
        self.modelo = modelo
        self.año = año
        
    def ficha_tecnica(self):
        """Retorna la ficha técnica básica del vehículo."""
        return f"Vehículo: {self.marca} {self.modelo} ({self.año})"

class Auto(Vehiculo):
    """Subclase concreta de Auto."""
    def ficha_tecnica(self):
        return super().ficha_tecnica() + " - Tipo: Auto"

class Moto(Vehiculo):
    """Subclase concreta de Moto."""
    def ficha_tecnica(self):
        return super().ficha_tecnica() + " - Tipo: Moto"

class AutoElectrico(Auto, Electrico):
    """Subclase con herencia múltiple de Auto y Electrico."""
    def __init__(self, marca, modelo, año, bateria_kwh):
        # Usamos kwargs para asegurar una correcta resolución en la herencia múltiple
        super().__init__(marca=marca, modelo=modelo, año=año, bateria_kwh=bateria_kwh)

    def ficha_tecnica(self):
        return f"{super().ficha_tecnica()} - Batería: {self.bateria_kwh} kWh"

# Verificando el MRO como lo pide la guía
auto_e = AutoElectrico("Tesla", "Model 3", 2023, 75)
print(auto_e.ficha_tecnica())
print(AutoElectrico.__mro__)
```

```python
#ejercicio 3
from abc import ABC, abstractmethod
import math

class Figura(ABC):
    """Clase abstracta base para figuras geométricas."""
    
    @abstractmethod
    def area(self):
        """Calcula el área de la figura."""
        pass

    @abstractmethod
    def perimetro(self):
        """Calcula el perímetro de la figura."""
        pass

    def describir(self):
        """Describe la figura utilizando sus métodos abstractos polimórficos."""
        return f"Área: {self.area():.2f}, Perímetro: {self.perimetro():.2f}"

class Circulo(Figura):
    def __init__(self, radio):
        self.radio = radio
        
    def area(self):
        return math.pi * (self.radio ** 2)
        
    def perimetro(self):
        return 2 * math.pi * self.radio

class Rectangulo(Figura):
    def __init__(self, base, altura):
        self.base = base
        self.altura = altura
        
    def area(self):
        return self.base * self.altura
        
    def perimetro(self):
        return 2 * (self.base + self.altura)

class Triangulo(Figura):
    def __init__(self, base, altura, hipotenusa):
        self.base = base
        self.altura = altura
        self.hipotenusa = hipotenusa
        
    def area(self):
        return (self.base * self.altura) / 2
        
    def perimetro(self):
        return self.base + self.altura + self.hipotenusa

def calcular_total(figuras: list):
    """Retorna la suma total de las áreas de una lista de figuras."""
    return sum(figura.area() for figura in figuras)

# Prueba del polimorfismo
figuras = [Circulo(5), Rectangulo(4, 6), Triangulo(3, 4, 5)]
print(f"Área total: {calcular_total(figuras):.2f}")
print(figuras[0].describir())
```

```python
#ejercicio 4
class MetodoPago(ABC):
    """Interfaz abstracta para métodos de pago."""
    
    @abstractmethod
    def procesar(self, monto):
        """Procesa el pago especificado."""
        pass

    @abstractmethod
    def reembolsar(self, monto):
        """Reembolsa el monto especificado."""
        pass

class PagoTarjeta(MetodoPago):
    def __init__(self, numero_tarjeta, titular):
        self.numero_tarjeta = numero_tarjeta
        self.titular = titular

    def procesar(self, monto):
        return f"Procesando ${monto} con tarjeta final {self.numero_tarjeta[-4:]} a nombre de {self.titular}"

    def reembolsar(self, monto):
        return f"Reembolsando ${monto} a la tarjeta de {self.titular}"

class PagoTransferencia(MetodoPago):
    def __init__(self, cuenta_origen, cuenta_destino):
        self.cuenta_origen = cuenta_origen
        self.cuenta_destino = cuenta_destino

    def procesar(self, monto):
        return f"Transfiriendo ${monto} de la cuenta {self.cuenta_origen} a la cuenta {self.cuenta_destino}"

    def reembolsar(self, monto):
        return f"Reembolsando transferencia de ${monto} a la cuenta {self.cuenta_origen}"

class PagoEfectivo(MetodoPago):
    def procesar(self, monto):
        return f"Procesando pago en efectivo por ${monto}"

    def reembolsar(self, monto):
        return f"Entregando reembolso físico en efectivo por ${monto}"

def ejecutar_pago(metodo: MetodoPago, monto: float):
    """Procesa un pago polimórfico imprimiendo su resultado en pantalla."""
    print(metodo.procesar(monto))

# Pruebas
ejecutar_pago(PagoTarjeta("1234567812349876", "Juan Pérez"), 50000)
ejecutar_pago(PagoEfectivo(), 15000)
```
