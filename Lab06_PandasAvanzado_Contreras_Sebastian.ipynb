```python
#Crea una función Python que reciba como parámetros: region, categoria y monto_minimo,
#y retorne el DataFrame filtrado según esos criterios. Pruébala con al menos 3 combinaciones
#diferentes.

def filtrar_ventas(region, categoria, monto_minimo):
    resultado = df[
        (df['region'] == region) & 
        (df['categoria'] == categoria) & 
        (df['monto_bruto'] >= monto_minimo)
    ]
    return resultado

# Pruebas con distintas combinaciones
prueba_1 = filtrar_ventas('Centro', 'Laptop', 5000)
prueba_2 = filtrar_ventas('Sur', 'Monitor', 1000)
prueba_3 = filtrar_ventas('Norte', 'Auriculares', 500)

print(f"Ventas encontradas Prueba 1: {prueba_1.shape[0]}")
print(f"Ventas encontradas Prueba 2: {prueba_2.shape[0]}")
print(f"Ventas encontradas Prueba 3: {prueba_3.shape[0]}")
```

```python
#Ejercicio 3 

# Agrupa el dataset por 'region' y calcula para la columna monto_bruto: suma total,
#promedio, máximo, mínimo y conteo de transacciones. Usa .agg() con nombres de
#columnas personalizados y aplica. round(2).

resumen_region = df.groupby('region').agg(
    suma_total=('monto_bruto', 'sum'),
    promedio=('monto_bruto', 'mean'),
    maximo=('monto_bruto', 'max'),
    minimo=('monto_bruto', 'min'),
    conteo=('monto_bruto', 'count')
).round(2) # 2 decimaless
```

```python
# 2. Sobre el resultado anterior, usa .reset_index() y ordena de mayor a menor según el total de
#ventas. Muestra el DataFrame resultante.

resumen_region = resumen_region.reset_index().sort_values(by='suma_total')

print(resumen_region)
```

```python
#3. Agrupa por ['region', 'mes'] y calcula la suma de monto_bruto y la suma de cantidad. Usa
#reset_index() al final. Muestra las primeras 10 filas

agrupacion = df.groupby(['region', 'mes'])[['monto_bruto', 'cantidad']].sum().reset_index()
print(agrupacion)
```

```python
#4. Identifica el vendedor con mayor suma de monto_bruto (ignorando nulos con dropna=False
#por defecto en groupby). Usa .groupby('vendedor')['monto_bruto'].sum().idxmax() para
#encontrar el nombre directamente

mejor_vendedor = df.groupby('vendedor', dropna=False)['monto_bruto'].sum().idxmax()

print(f"El vendedor con el mayor monto bruto de ventas es: {mejor_vendedor}")
```

```python
#5. Agrupa por categoría y calcula el precio_unit promedio y la cantidad total. Ordena por
#precio promedio descendente e interpreta el resultado: ¿que categoría tiene el mayor
#ticket promedio?

# 5. Agrupar por categoría: precio promedio y cantidad total
resumen_categoria = df.groupby('categoria').agg(precio_promedio=('precio_unit', 'mean'), cantidad_total=('cantidad', 'sum')).round(2).sort_values(by='precio_promedio', ascending=False)

print(resumen_categoria)
```
