```python
#Ejercicio 1
from functools import total_ordering

@total_ordering
class Producto:
    def __init__(self, nombre: str, precio: float, stock: int):
        self.nombre = nombre
        self.precio = float(precio)
        self.stock = int(stock)

    def __str__(self):
        precio_formateado = int(self.precio) if self.precio.is_integer() else self.precio
        return f"{self.nombre} ${precio_formateado} ({self.stock} unidades)"

    def __repr__(self):
        precio_formateado = int(self.precio) if self.precio.is_integer() else self.precio
        return f"Producto('{self.nombre}', {precio_formateado}, {self.stock})"

    def __eq__(self, otro):
        if not isinstance(otro, Producto):
            return NotImplemented
        return self.precio == otro.precio

    def __lt__(self, otro):
        if not isinstance(otro, Producto):
            return NotImplemented
        return self.precio < otro.precio

    def __add__(self, otro):
        if isinstance(otro, Producto) and self.nombre == otro.nombre:
            return Producto(self.nombre, self.precio, self.stock + otro.stock)
        raise ValueError("Solo se pueden sumar productos que tengan el mismo nombre.")

    def __mul__(self, n: float):
        #SE MULTIPLICA EL PRECIO
        return Producto(self.nombre, self.precio * n, self.stock)
```

```python
p1 = Producto('Arroz', 1200, 50)
p2 = Producto('Arroz', 900, 30)
p3 = Producto('Leche', 1200, 20)
print(p1) # Arroz – $1200 (50 unidades)
print(repr(p2)) # Producto('Arroz', 900, 30)
print(p1 == p3) # True → mismo precio
print(p2 < p1) # True → 900 < 1200
print(sorted([p1, p2, p3])) # ordenado por precio
combinado = p1 + p2 # stock = 80
print(combinado) # Arroz – $1200 (80 unidades)
oferta = p1 * 0.8 # precio con 20% descuento
print(oferta) # Arroz – $960.0 (50 unidades)
```

```python
#Ejercicio 2
class CuentaBancaria:
    def __init__(self, titular: str, saldo: float):
        self.titular = titular
        self.saldo = saldo

    @property
    def titular(self):
        return self._titular

    @titular.setter
    def titular(self, valor: str):
        if len(valor) < 3:
            raise ValueError("El nombre del titular debe tener al menos 3 caracteres.")
        self._titular = valor

    @property
    def saldo(self):
        return self._saldo

    @saldo.setter
    def saldo(self, valor: float):
        if valor < 0:
            raise ValueError("Saldo no puede ser negativo")
        self._saldo = valor

    def depositar(self, monto: float):
        if monto <= 0:
            raise ValueError("El monto a depositar debe ser mayor a 0.")
        self.saldo += monto 
#retiro
    def retirar(self, monto: float):
        if self.saldo - monto < 0:
            raise ValueError("No se puede realizar el retiro. El saldo quedaría negativo.")
        self.saldo -= monto # se usa el setter

    def __str__(self):
        return f"Cuenta de {self.titular} | Saldo: ${int(self.saldo)}"

    def __repr__(self):
        return f"CuentaBancaria(titular='{self.titular}', saldo={int(self.saldo)})"

    @property
    def aprobada(self):
        # SALDO SUPERIOR A 10.000 o igual
        return self.saldo >= 10000
```

```python
cuenta = CuentaBancaria('Ana López', 50000)
print(cuenta) # Cuenta de Ana López | Saldo: $50000
cuenta.depositar(20000)
print(cuenta.saldo) # 70000
cuenta.retirar(15000)
print(cuenta) # Cuenta de Ana López | Saldo: $55000
# Validaciones
try:
 cuenta.saldo = -1000 # debe lanzar ValueError
except ValueError as e:
 print(e) # Saldo no puede ser negativo
try:
 CuentaBancaria('AB', 0) # titular muy corto
except ValueError as e:
 print(e)
print(cuenta.aprobada) # True
```

```python
#Ejercicio 3

import math

class Figura:
    def __init__(self, color: str):
        self.color = color

    def area(self):
        raise NotImplementedError("El método area debe ser implementado en la subclase.")

    def perimetro(self):
        raise NotImplementedError("El método perimetro debe ser implementado en la subclase.")

    def __str__(self):
        try:
            val_area = self.area()
            val_perimetro = self.perimetro()
            
            area_str = f"{val_area:.2f}".rstrip('0').rstrip('.') if isinstance(val_area, float) else str(val_area)
            per_str = f"{val_perimetro:.2f}".rstrip('0').rstrip('.') if isinstance(val_perimetro, float) else str(val_perimetro)
            
            return f"Figura {self.color} | Área: {area_str} | Perímetro: {per_str}"
        except NotImplementedError:
            return f"Figura {self.color} (Abstracta)"

    def __repr__(self):
        return f"{self.__class__.__name__}('{self.color}')"

class Circulo(Figura):
    def __init__(self, color: str, radio: float):
        super().__init__(color)
        if radio <= 0:
            raise ValueError("El radio debe ser mayor a 0.")
        self.radio = radio

    def area(self):
        return math.pi * (self.radio ** 2)

    def perimetro(self):
        return 2 * math.pi * self.radio

class Rectangulo(Figura):
    def __init__(self, color: str, ancho: float, alto: float):
        super().__init__(color)
        if ancho <= 0 or alto <= 0:
            raise ValueError("El ancho y el alto deben ser mayores a 0.")
        self.ancho = ancho
        self.alto = alto
#anchoxalto
    def area(self):
        return self.ancho * self.alto
#perimetro
    def perimetro(self):
        return 2 * (self.ancho + self.alto)
#cuadrado
    @property
    def es_cuadrado(self):
        return self.ancho == self.alto
```

```python
c = Circulo('rojo', 5)
r = Rectangulo('azul', 4, 6)
cuad = Rectangulo('verde', 5, 5)
print(c) # Figura roja | Área: 78.54 | Perímetro: 31.42
print(r) # Figura azul | Área: 24 | Perímetro: 20
print(cuad.es_cuadrado) # True
print(r.es_cuadrado) # False
# Polimorfismo
figuras = [c, r, cuad]
for f in figuras:
 print(f'Área: {f.area():.2f}')
# Ordenar por área (requiere __lt__)
# Bonus: agrega __lt__ basado en área y ordénalas
```

```python
#Ejercicio 4
class Inventario:
    def __init__(self):
        #defino una lista
        self._productos = []

    def agregar(self, producto: Producto):
        self._productos.append(producto)

    def __len__(self):
        return len(self._productos)

    def __getitem__(self, i: int):
        return self._productos[i]

    def __contains__(self, producto: Producto):
        return producto in self._productos

    def __iter__(self):
        return iter(self._productos)

    def __str__(self):
        return "\n".join(str(p) for p in self._productos)

    def __repr__(self):
        return f"Inventario(n_productos={len(self._productos)})"

    @property
    def total_valor(self):
        return sum(p.precio * p.stock for p in self._productos)

    @property
    def mas_caro(self):
        if not self._productos:
            return None
        return max(self._productos)

    def buscar(self, nombre: str):
        nombre_lower = nombre.lower()
        return [p for p in self._productos if nombre_lower in p.nombre.lower()]
```

```python
inv = Inventario()
inv.agregar(Producto('Arroz', 1200, 50))
inv.agregar(Producto('Leche', 950, 30))
inv.agregar(Producto('Pan', 1800, 100))
inv.agregar(Producto('Panetón', 3500, 10))
print(len(inv)) # 4
print(inv[0]) # Arroz – $1200 (50 unidades)
p_buscar = Producto('Leche', 950, 30)
print(p_buscar in inv) # True
for p in inv:
 print(p)
print(inv.total_valor) # 1200*50 + 950*30 + 1800*100 + 3500*10 = ...
print(inv.mas_caro) # Panetón – $3500 (10 unidades)
resultados = inv.buscar('pan')
for r in resultados:
 print(r) # Pan y Panetón
```
