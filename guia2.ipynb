```python
from functools import reduce
import random
```

```python
#Actividad 2
#Ejercicio:
#Convertir velocidades de km/h a m/s usando map().
velocidades = [10, 20, 30, 40, 50]
conversion=list(map(lambda x: x/3.6, velocidades))
print("La conversión dada es", conversion)
#Conclusión: Se utilizó la conversión y expresión x/3.6 junto a las velocidades la dar con el resultado
#de las conversiones.
```

```python
#Actividad 3
temperaturas = [15, 18, 21, 35, 42, -10, 9, 30, 150]
#Filtrar valores válidos en el rango -20 a 120:
datos_validos = list(filter(lambda x: -20 <= x <= 120, temperaturas))
valores = [2,5,8,11,14,17,20]

valores_validos=list(filter(lambda x: x%2==0, valores))
print("Las temperaturas validas son:", valores_validos)
#Conclusión: Se validaron los datos con el list y filter, con la expresión -20<=x<=120 a los datos validos
#Purificados, se obtiene las temperaturas correctas.
```

```python
#Actividad 4
valores=[2,3,4,5]
resultado=reduce(lambda x,y: x*y,valores)
print("El resultado con respecto a los valores es:", resultado)
#Conclusión: Se utiliza la expresión x*y junto con la variable anonima lambda x,y para concluir el resultado de los
#valores.
```

```python
#Actividad 5
data = [12,15,22,35,41,8,3,55,62,90]
resultado = reduce(lambda x,y: x+y,
 map(lambda x: x*2,
 filter(lambda x: x>=10, data)))
print("El resultado es:", resultado)
#Conclusión: Junto con la lista data, se utiliza el reduce con variable x,y con el map de expresión x*2 y el filtrado
#x>=10 para entregar el valor exigido por la actividad.
```

```python
#Actividad 6
data = [random.randint(0,100) for _ in range(1000000)]

filtrado = list(filter(lambda x: x > 50, data))
resultado = len(filtrado)

cuadrado = list(map(lambda x: x**2, filtrado))

sumatotal = reduce(lambda x, y: x + y, cuadrado)
promedio = sumatotal / resultado

print("El promedio resultante es:", promedio)
#Conclusión: Se utiliza un list y filter para que la expresión "x>50" capture los datos de data correspondientes.
#Y con "resultado" leer con un len los datos filtrados.
#Se crea la función cuadrado para generar la expresión pedida.
#Lo mismo con sumatotal para crear la expresión de variable anonima "x+y"
#Y finalmente dar con el promedio de operación de sumatotal con resultado.
```

```python
#Desafio avanzado
sensores = [random.uniform(10,80) for _ in range(500000)]

procesados = list(map(lambda x: x + 273.15, filter(lambda x: 10 <= x <= 70, sensores)))
suma_total = reduce(lambda x, y: x + y, procesados)
promedio_kelvin= suma_total/len(procesados)
print("Temperatura promedio", promedio_kelvin, "°K")

#Conclusión: Se procesan los datos con el listado y un mapeo, con expresión "x+273.15" y con la condición 10<=x<=70
#Datos que se obtienen de sensores, para luego escribir la expresión reduce con "x+y" de los datos procesados anteriormente
#Para concluir finalmente con el promedio en graos kelvin y la impresión del resultado final con el promedio dado.
```
